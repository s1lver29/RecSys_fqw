```python
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_palette("muted")
sns.set_context("paper")
sns.set_style("darkgrid")
```

```python
mf = pd.read_excel("gmf_1m_4negatives_100negativtest.xlsx", index_col=0)
mf.head()
```

```python
mf["NDCG@10"].max()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.lineplot(data=mf, x="eposh", y="loss",
            ax=ax, label="Train loss")

sns.lineplot(data=mf, x="eposh", y="test_loss",
             ax=ax, label="Test loss")

ax.set_title("Matrix Factorization")
ax.set_ylabel("Loss")
ax.set_xlabel("Epoch")
ax.set_xticks(range(1, 51))

plt.legend(loc="upper right")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.lineplot(data=mf, x="eposh", y="HR@10",
            ax=ax, label="HR@10")

sns.lineplot(data=mf, x="eposh", y="Recall@10",
             ax=ax, label="Recall@10")

sns.lineplot(data=mf, x="eposh", y="MAP@10",
             ax=ax, label="MAP@10")

sns.lineplot(data=mf, x="eposh", y="NDCG@10",
             ax=ax, label="NDCG@10")

sns.lineplot(data=mf, x="eposh", y="MRR@10",
             ax=ax, label="MRR@10")

ax.set_title("Matrix Factorization")
ax.set_ylabel("Loss")
ax.set_xlabel("Epoch")
ax.set_xticks(range(1, 51))

plt.legend(loc="upper right")
plt.show()
```
